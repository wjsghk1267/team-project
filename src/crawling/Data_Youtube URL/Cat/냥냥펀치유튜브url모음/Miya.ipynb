```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# ChromeDriver ì´ˆê¸°í™”
driver = webdriver.Chrome()

# í•´ë‹¹ í˜ì´ì§€ë¡œ ì´ë™
driver.get('https://www.youtube.com/playlist?list=PLhLLGKcG94RMTaxlgxm7Vgr76ca3Zv5M0')

# í˜ì´ì§€ê°€ ë¡œë“œë  ì‹œê°„ì„ ê¸°ë‹¤ë¦¼
time.sleep(3)

# í˜ì´ì§€ ëê¹Œì§€ ìŠ¤í¬ë¡¤ì„ ë‚´ë¦¬ëŠ” í•¨ìˆ˜
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    while True:
        # í˜ì´ì§€ì˜ ë§¨ ì•„ë˜ê¹Œì§€ ìŠ¤í¬ë¡¤
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        
        # ìƒˆë¡œìš´ ì½˜í…ì¸ ê°€ ë¡œë“œë  ì‹œê°„ì„ ê¸°ë‹¤ë¦¼
        time.sleep(2)

        # ìŠ¤í¬ë¡¤ í›„ ìƒˆë¡œìš´ ë†’ì´ë¥¼ í™•ì¸
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        
        # ë” ì´ìƒ ìƒˆë¡œìš´ ì½˜í…ì¸ ê°€ ì—†ìœ¼ë©´ ìŠ¤í¬ë¡¤ì„ ë©ˆì¶¤
        if new_height == last_height:
            break
        
        last_height = new_height

# í˜ì´ì§€ ëê¹Œì§€ ìŠ¤í¬ë¡¤
scroll_to_bottom(driver)

# ëª¨ë“  video-title ìš”ì†Œë¥¼ ê°€ì ¸ì˜´
video_titles = driver.find_elements(By.XPATH, '//*[@id="video-title"]')

# ë°ì´í„°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
data = []

# ê° ì˜ìƒ ì œëª©ê³¼ hrefë¥¼ ì¶”ì¶œí•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
for video in video_titles:
    title = video.get_attribute('title')  # ì œëª©ì„ ê°€ì ¸ì˜´
    href = video.get_attribute('href')    # href ë§í¬ë¥¼ ê°€ì ¸ì˜´
    data.append({'ì œëª©': title, 'ë§í¬': href})

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
df = pd.DataFrame(data)

# CSV íŒŒì¼ë¡œ ì €ì¥
df.to_csv('ğŸ‘¨â€âš•ï¸ë¬˜í•œ ì§„ë£Œì‹¤.csv', index=False, encoding='utf-8-sig')

# ë“œë¼ì´ë²„ ì¢…ë£Œ
driver.quit()
```
